```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
genislik = 200
yukseklik = 200
kanal_sayisi = 3

resim_array = np.random.randint(0, 256, (yukseklik, genislik, kanal_sayisi), dtype=np.uint8)

plt.imshow(resim_array)
plt.axis('off')  # Eksenleri kapat
plt.show()
```

```python
resim_array=np.resize(resim_array,(150,150,3))# resmi tekrardan boyutlandrdk

plt.imshow(resim_array)
plt.axis('off')  # Eksenleri kapat
plt.show()
```

```python
gray_image = np.dot(resim_array[...,:3], [0.2989, 0.5870, 0.1140])
#grselin tm piksellerin ilk  kanaln (Krmz, Yeil, Mavi) seer.
#Sa taraftaki diziyle arpm ilemini gerekletirir.
plt.imshow(gray_image)
plt.axis('off')  # Eksenleri kapat
plt.show()
```

```python
cropped_image = resim_array[10:20,20:30]

plt.imshow(cropped_image)
plt.axis('off')  # Eksenleri kapat
plt.show()

```

```python
for i in range(10, 100):
    for j in range(20, 100):
        resim_array[i, j] = np.array([255, 255, 255])
        
        
plt.imshow(resim_array)
plt.axis('off')  # Eksenleri kapat
plt.show()

    
```

```python
x = np.arange(0,150)
y = np.arange(0, 150, 10)

for i in x:
    for j in y:
        resim_array[j,i]=np.array([255, 255, 255])
        
plt.imshow(resim_array)
plt.axis('off')  # Eksenleri kapat
plt.show()
        
```
